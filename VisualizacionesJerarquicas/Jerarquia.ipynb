```python
import pandas as pd
import plotly.express as px
```

```python
dfpaises=pd.read_csv('https://raw.githubusercontent.com/lromeror/DataVisualization/main/VisualMultiVariables/paises_vida.csv')
dfpaises=dfpaises.dropna()
df2007=dfpaises[dfpaises["year"]==2007]
df2007
```

```python
fig=px.treemap(df2007,
               path=[px.Constant('World'),'continent','country'],
               values='pop',
               color_continuous_scale='RdBu',
               color='lifeExp')
fig.show()
```
