```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
df_edges=pd.read_csv('https://raw.githubusercontent.com/lromeror/DataVisualization/main/VisualizacionGraph/edges.csv')
df_nodes=pd.read_csv('https://raw.githubusercontent.com/lromeror/DataVisualization/main/VisualizacionGraph/nodes.csv')
```

##Grafos


- Relativamente pocos puntos se vuelven muy densas

```python
df_edges.head()
```

```python
df_nodes.head()
```

```python
G=nx.from_pandas_edgelist(df_edges, source='src', target='dst', edge_attr='weight')
G.nodes()
```

```python
plt.rcParams['figure.figsize'] = [10, 7]
nx.draw(G)
```

```python
nx.draw_networkx(G)
```

```python
nx.draw_circular(G)
```

```python
nx.draw_spring(G,with_labels=True,font_size=10,width=list(df_edges['weight']/5)
               ,style=":",node_size=list(df_nodes['pagerank']*200),
               node_color=list(df_nodes['community']),cmap='tab10'
               ,nodelist=list(df_nodes['id']))
```
