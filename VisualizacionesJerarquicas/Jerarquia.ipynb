```python
import pandas as pd
import plotly.express as px
```

```python
dfpaises=pd.read_csv('https://raw.githubusercontent.com/lromeror/DataVisualization/main/VisualMultiVariables/paises_vida.csv')
dfpaises=dfpaises.dropna()
df2007=dfpaises[dfpaises["year"]==2007]
df2007
```

```python
fig=px.treemap(df2007,
               path=[px.Constant('World'),'continent','country'],
               values='pop',
               color_continuous_scale='RdBu',
               color='lifeExp')
fig.show()
```

```python
dfmen=pd.read_csv('https://raw.githubusercontent.com/lromeror/DataVisualization/main/VisualizacionesJerarquicas/marcas_mensajes.csv')
dfmen
```

```python
fig=px.treemap(dfmen,
               path=[px.Constant('All'),'tipo','marca'],
               values='n_mensajes')
fig.data[0].textinfo='label+text+value'
fig.show()
```

## Sunburst
- mejor para ver proporciones
- diferencia entre proporciones

```python
fig = px.sunburst(
    df2007,
    path=[px.Constant("world"), 'continent', 'country'],
    values='pop',
    color='lifeExp', # Codificamos una segunda variable numÃ©rica utilizando colores
    color_continuous_scale='RdBu',
)
fig.show()
```

```python
dfmar=pd.read_csv('https://raw.githubusercontent.com/lromeror/DataVisualization/main/VisualizacionesJerarquicas/marcas_mensajes.csv')
dfmar
```
