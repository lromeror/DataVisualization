#Limpieza de datos

```python
#df_filtrado=df.dropna() #Esto elimina las filas Nan
df_filtrado=df.fillna({'retweets':0,'mentions':-1}) #Reempalzar datos con valor especifico
```

```python
def calculateGanancias(retwet_num):
  ganancia=retwet_num*rd.randint(3,5)
  return ganancia
```

```python
df_filtrado['ganancias']=df['retweets'].apply(calculateGanancias)#apply dice que a la columna le va aplicar la funcion previamente definida "calculateGanancias"
```

```python
df_filtrado.head()
```

```python
#forma 1 con funciones
def calculatepo(fila):
  resultado =fila['followees']/fila['followers']
  return resultado
```

```python
df_filtrado['popularity']=df_filtrado.apply(calculatepo,axis=1)#columna es 0 por fila es 1
```

```python
df_filtrado['popularity2']=df_filtrado['followees']/df_filtrado['followers']#forma nuemro 2 de hacer lo mismo
```
