```python
import pandas as pd
```

```python
df1= pd.read_csv('https://raw.githubusercontent.com/lromeror/DataVisualization/main/Joins%20and%20DateTime/dataset_2a.csv')
df2= pd.read_csv('https://raw.githubusercontent.com/lromeror/DataVisualization/main/Joins%20and%20DateTime/dataset_2b.csv')
df3= pd.read_csv('https://raw.githubusercontent.com/lromeror/DataVisualization/main/Joins%20and%20DateTime/dataset_2c.csv')
```

#Join dataset

```python
df_with_user=df1.merge(df2, on='user', how='left')#join como el de basedatos
df_with_user
```

```python
df_full=df_with_user.merge(df3, on='ciudad', how='left')
df_full
```

#DateTime

```python
df_full['fecha']=pd.to_datetime(df_full['fecha_publicado'],format='%d/%m/%Y')
df_full
```

```python
df_full['mes']=df_full['fecha_publicado'].apply(lambda x:x[3:])
df_full
```

```python
#Para calcular la cantidad de Twitt que se trabajo en un mes
group=df_full.groupby(['mes','ciudad']).agg({'id':'count'})
```

```python
group=group.reset_index([1])
group
```

Pivotar para convertir cada valor unico de una columna en nuevas columnas

```python
group_pivot=group.pivot(columns='ciudad')
group_pivot
group_pivot=group_pivot.droplevel(0, axis='columns')
#Las columnas tambien pueden tener mas de 2 niveles
```

```python
group_pivot=group_pivot.fillna(0)
group_pivot
```

```python
group_pivot=group_pivot.transpose()
group_pivot
```
